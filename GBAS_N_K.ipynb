```python
# Stopping Rule Algorithm (SRA)
def GBAS(k, samples):
    count = 0 # the counting variable of GBAS
    rexp = 0 # the random variable that follows exponential distribution
    # Visit the samples 
    i = 0
    while i < len(samples):
        sample_i = samples[i]
        rexp += np.random.exponential(scale = 1.0, size = 1)[0]
        i += 1
        # Generate a random variable u over [0, 1], estimate the probability that u<= sample_i 
        u = np.random.uniform(low = 0.0, high = 1.0, size = 1)[0]
        if u <= sample_i:
            count += 1
        if count == k:
            miu = (k-1)/rexp # the reliability value
            return (miu, i)
    return (-1, i)
```

```python
def average_fl(nk):
    nk_fl = {}
    for key, value in nk.items():
        nk_fl[key] = sum(value.values()) / 23
    return list(nk_fl.values()) 
```
